```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

from uncertainty_forest.uncertainty_forest import UncertaintyForest
from sklearn.ensemble import RandomForestClassifier
from tqdm import tqdm_notebook
from joblib import Parallel, delayed
from scipy.stats import entropy, norm
```

```python
def generate_data(n, d, frac_eval, mu = 1, sigma = 1):
    n_1 = np.random.binomial(n, .5) # number of class 1
    mean = np.zeros(d)
    mean[0] = mu
    X_1 = np.random.multivariate_normal(mean, sigma * np.eye(d), n_1)
    
    X = np.concatenate((X_1, np.random.multivariate_normal(mean, sigma * np.eye(d), n - n_1)))
    y = np.concatenate((np.repeat(1, n_1), np.repeat(-1, n - n_1)))
    
    # Evaluation data.
    n_eval = int(np.floor(frac_eval*n))
    eval_indices = np.random.choice(np.arange(n), size = n_eval, replace = False)
    X_eval = X[eval_indices, :]
    X = np.delete(X, eval_indices, axis = 0)
    y = np.delete(y, eval_indices, axis = 0)
    return X, y, X_eval
```

```python
def conditional_entropy_distribution(n, d, frac_eval, algos, num_trials, parallel = False):
    # For each trial, generate data and compute conditional entropy for each algorithm.
    def worker(t):
        X, y, X_eval = generate_data(n, d, frac_eval)
        ret = np.zeros(len(algos))
        for i in range(len(algos)):
            obj = algos[i]['instance']
            obj.fit(X, y)
            p = obj.predict_proba(X_eval)
            ret[i] = np.mean(entropy(p.T, base = 2))
        return ret
    
    if parallel:
        predicted_cond_entropy = np.array(Parallel(n_jobs=-2)(delayed(worker)(t) for t in range(num_trials)))
    else:
        predicted_cond_entropy = np.zeros((num_trials, len(algos)))
        for t in tqdm_notebook(range(num_trials)):
            predicted_cond_entropy[t, :] = worker(t)
            
    return predicted_cond_entropy
```

```python
def conditional_entropy_by_n(sample_sizes, d, frac_eval, algos, num_trials, parallel = False):
    # Repeat for all 'n', and save output in the 'algos' array.
    cond_entropy_range = np.zeros((len(sample_sizes), num_trials, len(algos)))
    for i in range(len(sample_sizes)):
        cond_entropy_range[i, :, :] = conditional_entropy_distribution(sample_sizes[i], d, frac_eval, algos, num_trials)
        
    for j in range(len(algos)):
        algos[j]['cond_entropy_range'] = cond_entropy_range[:, :, j]
        
    with open('algos_fig2.pkl', 'wb') as f:
        pickle.dump(algos, f)
    with open('sample_sizes_fig2.pkl', 'wb') as f:
        pickle.dump(sample_sizes, f)
```

```python
def plot_cond_entropy_range(sample_sizes, algos, num_plotted_trials, d):
    sns.set(font_scale = 2)
    sns.set_style("ticks")
    plt.figure()
    plt.rcParams["font.family"] = "sans-serif"
    plt.rcParams['figure.figsize'] = [15, 10]
    
    for algo in algos:
        # Plot the mean over trials as a solid line.
        plt.plot(sample_sizes, 
                 np.mean(algo['cond_entropy_range'], axis = 1).flatten(), 
                 label = algo['label'], 
                 linewidth = 2, 
                 color = algo['color'])
        # Use transparent lines to show other trials.
        for t in range(num_plotted_trials):
            plt.plot(sample_sizes, 
                     algo['cond_entropy_range'][:, t].flatten(),  
                     linewidth = 4, 
                     color = algo['color'],
                     alpha = 0.15)
    
    truth = .356
    plt.axhline(y = truth, linestyle = '-', color = "black", label = "Truth")
            
    plt.xlabel("Sample Size")
    plt.ylabel("Estimated Conditional Entropy")
    plt.title("Conditional Entropy versus Sample Size, d = %d" % d)
    plt.ylim(ymin = -0.05)
    plt.yticks(fontsize = 20)
    plt.xticks(fontsize = 20)
    plt.legend(loc = 'upper right')
    plt.tight_layout()
    plt.savefig("fig2_dim_%d.pdf" % d)
    plt.show()
```

```python
def compute_X_entropy(mu, xmin, xmax, step):
    x_range = range(xmin, xmax, step)
    def worker(x):
        p = 0.5 * norm.pdf(x, loc = mu, scale = 1) + 0.5 * norm.pdf(x, loc = -mu, scale = 1)
        return p * np.log2(p) * step
    return np.sum(Parallel(n_jobs=-2)(delayed(worker)(x) for x in x_range))
```

```python
mu = 1
xmin = -1000
xmax = 1000
step = 1

print(compute_X_entropy(mu, xmin, xmax, step))
```

```python
# Data.
sample_sizes = range(10, 21, 5)
d = 1
num_trials = 4
frac_eval = 0.3

# Algorithms.
algos = [
    {
        'instance': RandomForestClassifier(min_samples_leaf = 6, n_estimators = 30),
        'label': 'CART',
        'title': 'CART Forest',
        'color': "#1b9e77",
    },
    {
        'instance': UncertaintyForest(finite_correction = False, min_samples_leaf = 6, n_estimators = 30),
        'label': 'Honest',
        'title': 'Honest Forest',
        'color': "#fdae61",
    },
    {
        'instance': UncertaintyForest(min_samples_leaf = 6, n_estimators = 30),
        'label': 'UF',
        'title': 'Uncertainty Forest',
        'color': "#F41711",
    },
]

# Plotting parameters.
num_plotted_trials = 3
```

```python
# Estimate conditional entropy.
conditional_entropy_by_n(sample_sizes, d, frac_eval, algos, num_trials, parallel = True)
print(algos[0]['cond_entropy_range'])
```

```python
with open('sample_sizes_fig2.pkl', 'rb') as f:
    sample_sizes = pickle.load(f)
with open('algos_fig2.pkl', 'rb') as f:
    algos = pickle.load(f)
    
plot_cond_entropy_range(sample_sizes, algos, num_plotted_trials, d)
```
